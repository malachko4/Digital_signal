## 1. Методами порогової та морфологічної обробки відобразити тільки контури циферблату
 Завантажити зображення "digit.jpg" директорії "./examples/"
 


```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread("./digit.jpg", 2)
plt.imshow(img, cmap = 'gray', vmin = 0, vmax = 255);
plt.show();
```

```python
ret, th = cv2.threshold(img, 0, 255, cv2.THRESH_OTSU)

plt.figure(figsize=(20,20))
plt.subplot(236),plt.imshow(th, 'gray')
plt.title(f"Otsu's Thresholding ret2 = {str(ret)}"), plt.xticks([]), plt.yticks([])

plt.show()
```

## 2. Методами порогової та морфологічної обробки відобразити границі баскетбольного поля
 Завантажити зображення "basketball.jpg" директорії "./examples/"
 


```python
img = cv2.imread("./basketball.jpg", cv2.IMREAD_GRAYSCALE)
plt.imshow(img, cmap = 'gray', vmin = 0, vmax = 255);
plt.show();
```

```python
ret,th1 = cv2.threshold(img,180,200,cv2.THRESH_BINARY)

kernel = np.ones((5,5),np.uint8);
opening = cv2.morphologyEx(th1 , cv2.MORPH_OPEN, kernel);

plt.figure(figsize=(20,20));
plt.subplot(122);plt.imshow(opening , cmap = 'gray'); plt.title('Global Thresholding (v = 127)');
plt.show();
```
